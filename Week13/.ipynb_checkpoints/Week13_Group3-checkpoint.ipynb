**soultion:**  
1.
$$
\begin{aligned}
\frac{\partial}{\partial x}\psi(x,t)=\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty}F^{-1}[\psi]ike^{ikx}dk
\end{aligned}
$$
所以先求$\psi$的逆傅里叶变换，乘$ik$再做傅里叶变换就能得到导数了

````python
#by Zhang Xi

import numpy as np
import matplotlib.pyplot as plt

f = lambda x: (1/(4*np.pi))**0.25 * np.exp(-x**2/2)


def FFT(f):    #快速傅里叶
    # Compute all 2-point DFTs
    N = f.shape[0]
    W = np.array([[1, 1], [1, -1]])
    F = np.dot(W, f.reshape([2, -1]))

    # number of remaining Cooley-Tukey stages
    stages = int(np.log2(N))-1
    for i in range(stages):
        k = F.shape[0] # size of the DFTs to combine
        n = F.shape[1] # number of DFTs to combine
        Am = F[:, :n//2] # 'even' terms
        Bm = F[:, n//2:] # 'odd' terms
        twiddle = np.exp(-1.j*np.pi*np.arange(k)/k)[:, None]
        F = np.vstack([Am + twiddle*Bm, Am - twiddle*Bm])
    return F.flatten()

def FFT_inv(f):    #快速逆傅里叶
    N = f.shape[0]
    W = np.array([[1, 1], [1, -1]])
    F = np.dot(W, f.reshape([2, -1]))

    # number of remaining Cooley-Tukey stages
    stages = int(np.log2(N))-1
    for i in range(stages):
        k = F.shape[0] # size of the DFTs to combine
        n = F.shape[1] # number of DFTs to combine
        Am = F[:, :n//2] # 'even' terms
        Bm = F[:, n//2:] # 'odd' terms
        twiddle = np.exp(1.j*np.pi*np.arange(k)/k)[:, None]
        F = np.vstack([Am + twiddle*Bm, Am - twiddle*Bm])
    return F.flatten()/len(f)


def Fourier_differentiation_prime(f,x):   #傅里叶求导数
    N = len(x)
    a = (N-1)/2
    dx = (x[-1] - x[0])/( N - 1 )
    dk = 2*np.pi/(x[-1] - x[0])
    #psi to phi\
    j = np.complex(0,1.0)
    n = np.arange(N)
    shift1 = np.exp(2*np.pi*j*(n-a)*a/N)
    shift2 = np.exp(-2*np.pi*j*(n-a)*a/N)
    
    phi = shift1 * FFT( f*np.exp(2*np.pi*j*a*n/N) )   #利用离散傅里叶变换求傅里叶变换
    psi = shift2 * FFT_inv(j*(n-a)*dk*phi*np.exp(-2*np.pi*j*a*n/N))  #乘ik后逆变换
    return np.real(psi)
    
N = 256
a = (N - 1 ) / 2
halfL = 5
dx = (halfL - (-halfL)) / (N - 1)

# define the x-grid and discretised wavefunction
x = np.arange(-halfL, halfL+dx, dx)
psi = f(x)
plt.figure(dpi = 160)
plt.plot(x,np.real(Fourier_differentiation_prime(psi,x)))
plt.xlabel('x')
plt.ylabel(r'$\psi^{\prime}$')
plt.show()
````

````python
#二阶导就是求两次导数，调用函数两次
psi_prime = Fourier_differentiation_prime(psi,x)
psi_prime2 = Fourier_differentiation_prime(psi_prime,x)

plt.figure(dpi=160)
plt.plot(x,psi_prime2)
plt.xlabel('x')
plt.ylabel(r'$\psi^{\prime\prime}$')
plt.show()
````

````python
x_max = 5
f_prime = lambda x: -(1/(4*np.pi))**0.25*x*np.exp(-x**2/2)
f_prime2 = lambda x: (x*x-1)*(1/(4*np.pi))**0.25*np.exp(-x**2/2)

ii = np.array([2,3,4,5,6,7,8,9,10])
errormax = np.zeros((len(ii),2))
count = 0

x = np.linspace(-x_max,x_max,256)

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(16,5))

psi_prime = Fourier_differentiation_prime(f(x),x)
ax1.plot(x,psi_prime,label='Calculated')
ax1.plot(x,f_prime(x),'-.',label='Exact value')
ax1.set_xlabel('x',fontsize=14)
ax1.set_ylabel(r'$\psi^{\prime}$',fontsize=14)
ax1.legend(fontsize=14)

ax2.plot(x,Fourier_differentiation_prime(psi_prime,x),label='Calculated')
ax2.plot(x,f_prime2(x),'-.',label='Exact value')
ax2.set_xlabel('x',fontsize=14)
ax2.set_ylabel(r'$\psi^{\prime\prime}$',fontsize=14)
ax2.legend(fontsize=14)

plt.show()
plt.close()

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(16,5))
E_r1 = (psi_prime - f_prime(x))/psi_prime
ax1.plot(x,E_r1)
ax1.set_xlabel('x',fontsize=14)
ax1.set_ylabel(r'Relative Error',fontsize=14)

E_r2 = (Fourier_differentiation_prime(psi_prime,x) - f_prime2(x))/f_prime2(x)
ax2.plot(x,E_r2)
ax2.set_xlabel('x',fontsize=14)
ax2.set_ylabel(r'Relative Error',fontsize=14)

plt.show()
plt.close()

for i in ii:
    N = 2**i
    x = np.linspace(-x_max,x_max,N)
    psi = f(x)
    psi_prime_exact = f_prime(x)
    psi_prime2_exact = f_prime2(x)
    psi_prime_cal = Fourier_differentiation_prime(psi,x)
    psi_prime2_cal = Fourier_differentiation_prime(psi_prime_cal,x)
    errormax[count,:] = np.array([np.max(abs(psi_prime_cal-psi_prime_exact)),np.max(abs(psi_prime2_cal - psi_prime2_exact))])
    count += 1
    
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(16,5))
xaxis = np.log(2**ii)
ax1.plot(xaxis,np.log(errormax[:,0]),'o')
ax1.set_xlabel(r'$\ln{N}$',fontsize = 14)
ax1.set_ylabel(r'$\ln{E_{max}}$',fontsize=14)
z = np.polyfit(xaxis,np.log(errormax[:,0]),1)
ax1.plot(xaxis,z[0]*xaxis+z[1],label = 'y='+str(f'{z[0]:.6}')+'x'+str(f'{z[1]:.6}'))
ax1.set_title(r'Max Error of $\psi^{\prime}$',fontsize = 14)
ax1.legend()

ax2.plot(np.log(2**ii),np.log(errormax[:,1]),'o')

z2 = np.polyfit(xaxis[:-2],np.log(errormax[:,1])[:-2],1)
ax2.plot(xaxis[0:-2],z2[0]*xaxis[0:-2]+z2[1],label = 'y='+str(f'{z2[0]:.6}')+'x'+str(f'{z2[1]:.6}'))
ax2.set_xlabel(r'$\ln{N}$',fontsize = 14)
ax2.set_ylabel(r'$\ln{E_{max}}$',fontsize=14)
ax2.set_title(r'Max Error of $\psi^{\prime\prime}$',fontsize = 14)
ax2.legend()

plt.show()
````

分析能量$$\begin{aligned}
E_{n+1}&=mgx_{n+1}+\frac{1}{2}mv_{n+1}^2\\
&=mg(x_n+v_ndt)+\frac{1}{2}m(v_n-gdt)^2\\
&=E_n+\frac{1}{2}mg^2dt^2
\end{aligned}
$$
每更新一次增加$\frac{1}{2}mg^2dt^2$的能量，总时间T不变的情况下，要会更新$T/dT$次，能量增加$\frac{1}{2}mg^2Tdt\propto dt$

````python
N = 300
x = np.zeros(N)
v = np.zeros(N)
t = np.zeros(N)

def init_cond(x,v,t,x0,v0):
    x[0] = x0 # initial height of the ball
    v[0] = v0 # initial velocity of the ball
    g = 9.8 # gravitational acceleration
    t[0] = 0 # initial time
    dt = 0.01 # size of time step
    # loop for 300 timesteps
    for step in range(N-1):
        t[step+1] = t[step] + dt
        v[step+1] = v[step] - g*dt
        x[step+1] = x[step] + v[step+1]*dt

        # reflect the motion of the ball
        # when it hits the surface x=0
        if x[step+1] < 0:
            x[step+1] = -x[step+1]
            v[step+1] = -v[step+1]
    return x,v,t

for x_init in [1.0,2.0]:
    for v_init in [0.0,5.0]:
        x,v,t = init_cond(x,v,t,x_init,v_init)

        plt.plot(t,x)
        plt.xlabel('time t')
        plt.ylabel('position x')
        plt.title(f'x0 = {x_init}, v0 = {v_init}')
        plt.show()
        plt.close()
````
