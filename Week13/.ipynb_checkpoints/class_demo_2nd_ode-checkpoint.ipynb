$y^{\prime^\prime} = -\omega^2y$   
$t\in [0,5\pi]$
$y(t=0)=0, y^\prime(t=0)=1$

```python
# define drive function
def f(y, t, omega):
    pos, velocity = y
    dydt = [velocity, -omega**2 * pos]
    return dydt
```

```python
# generage t grid
dt = 0.1
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.arange(0, 5*np.pi+dt, dt)
```

```python
# initialize y grid
y = np.zeros((len(t),2))
```

```python
y.shape
```

```python
# initial condition
y[0]  = np.array([0,1])
```

```python
# forward euler algorithm
omega = 1
for n in range(0, len(t) - 1):
    y[n+1] = y[n] + dt * np.array(f(y[n], t[n], omega))
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,7))
ax[0].plot(t,y[:,0],'ro-')
ax[0].set_xlabel('t')
ax[0].set_ylabel('Position')

ax[1].plot(t,y[:,1],'bs-')
ax[0].set_xlabel('t')
ax[0].set_ylabel('Velocity')
```

```python
# euler-cromer algorithm
omega = 1
for n in range(0, len(t) - 1):
    y[n+1,0] = y[n,0] + dt * y[n,1]
    y[n+1,1] = y[n,1] + dt * (-omega**2 * y[n+1,0])
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,7))
ax[0].plot(t,y[:,0],'ro-')
ax[0].set_xlabel('t')
ax[0].set_ylabel('Position')

ax[1].plot(t,y[:,1],'bs-')
ax[0].set_xlabel('t')
ax[0].set_ylabel('Velocity')
```

```python
# RK4
#...
y_rk4 = np.zeros((len(t),2))
y_rk4[0] = np.array([0,1])
for n in range(0, len(t) - 1):
    k1 = np.array(f(y[n], t[n], omega))
    k2 = np.array(f(y[n] + 0.5*dt*k1, t[n] + 0.5*dt, omega))
    k3 = np.array(f(y[n] + 0.5*dt*k2, t[n] + 0.5*dt, omega))
    k4 = np.array(f(y[n] + dt*k3, t[n] + dt, omega))
    y_rk4[n+1] = y_rk4[n] + dt * (k1 +  2 * k2 + 2 * k3 + k4)/6
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,7))
ax[0].plot(t,y_rk4[:,0],'ro-')
ax[0].set_xlabel('t')
ax[0].set_ylabel('Position')

ax[1].plot(t,y_rk4[:,1],'bs-')
ax[1].set_xlabel('t')
ax[1].set_ylabel('Velocity')
```
